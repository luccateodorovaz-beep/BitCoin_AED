### 1) Importação e limpeza dos dados do Bitcoin

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = '../data/bitcoin_dataset.csv' # Arquivo que contém os dados do bicoin

df = pd.read_csv(data_path)

df['Date'] = pd.to_datetime(df['Date']) # Lembara de comentar

df.head()
```

```python
sns.set_theme(style='darkgrid')

plt.figure(figsize=(12,6))

sns.lineplot(x='Date', y='Close', data=df)

plt.title('Evolução do Preço do Bitcoin (BTC)', fontsize=16)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Preço (USD)', fontsize=12)

plt.show()
```
